### Numerical evaluation of $ \frac{d}{dx}[cos(x)] $

```python
import numpy as np

def deriv_fwd(f, x, dx):
    return (f(x+dx)-f(x))/dx

def deriv_cen(f, x, dx):
    return (f(x+dx/2)-f(x-dx/2))/dx

x = np.pi/2
dx = 0.00001

print("Derivative of cos(x) at x = {}".format(x))
print("Actual Value:", -1*np.sin(x))
print("Forward Step:", deriv_fwd(np.cos, x, dx))
print("Center Step: ", deriv_cen(np.cos, x, dx))
```

### 1) Error growth - Forward Difference

```python
import numpy as np
import matplotlib.pyplot as plt

x = []
y = []

dx = 1e-16
for i in range(16):
    x.append(dx)
    y.append(np.fabs(deriv_fwd(np.cos, np.pi/4, dx)+np.sin(np.pi/4)))
    dx*=10

plt.loglog(x, y, 'b-o')
plt.title("Forward Difference Error", fontsize=24)
plt.xlabel("dx", fontsize=18)
plt.ylabel("Error", fontsize=18)
plt.grid("True")
plt.show()
```

<p> Using the forward difference method, the error grows at approximately $ O(dx^2) $. The error is at its lowest at a dx value of $ ~10^{-8} $. </p>

### 2) Error growth - Central Difference

```python
import numpy as np
import matplotlib.pyplot as plt

x = []
y = []

dx = 1e-16
for i in range(16):
    x.append(dx)
    y.append(np.fabs(deriv_cen(np.cos, np.pi/4, dx)+np.sin(np.pi/4)))
    dx*=10

plt.loglog(x, y, 'b-o')
plt.title("Central Difference Error", fontsize=24)
plt.xlabel("dx", fontsize=18)
plt.ylabel("Error", fontsize=18)
plt.grid("True")
plt.show()
```

<p> Using the central difference method, the error drops as $ O(dx^2) $, which is the same as the forward difference method. However, the minimal error occurs at $ dx=10^{-5} $, which is three orders of magnitude larger than the forward difference method.</p>
<p> In addition, the error drops to a minimum of $ 10^{-11} $, compared to the minimum of $ 10^{-8} $ of the forward difference method. This shows that the central difference method is more precise with a larger step size. </p>

### 3) Central Difference at x = 0

```python
import numpy as np
import matplotlib.pyplot as plt

x = []
y = []

dx = 1e-16
for i in range(16):
    x.append(dx)
    y.append(np.fabs(deriv_cen(np.cos, 0, dx)+np.sin(0)))
    dx*=10

#Use a semi-log plot instead of a log-log plot.
plt.semilogx(x, y, 'b-o')
plt.xlabel("dx", fontsize=18)
plt.ylabel("Error", fontsize=18)
plt.grid("True")
plt.show()
```

<p> The "errors" calculated at x=0 are all 0. This is because $cos(x)$ is symmetric about x=0, and </p>
<p> \begin{equation*}
cos(x)=cos(-x)
\end{equation*} </p>
<p> Therefore, the central difference approximation will always return a slope of zero. </p>
